### User-Based Recommendation

```python
import pandas as pd
```

```python
movie = pd.read_csv('/Users/yelizsari/Downloads/VBO/6. VBO - Recommendation/movie.csv')
rating = pd.read_csv('/Users/yelizsari/Downloads/VBO/6. VBO - Recommendation/rating.csv')
df = movie.merge(rating, how='left', on='movieId')
```

```python
rating_counts = pd.DataFrame(df.title.value_counts())
rare_movies = rating_counts[rating_counts['title'] <= 1000].index
common_movies = df[~df['title'].isin(rare_movies)]
```

```python
user_movie_df = common_movies.pivot_table(index = ["userId"], columns = ["title"], values = "rating")
```

```python
user_movie_df.head()
```

```python
random_user_df = user_movie_df[user_movie_df.index == 108170]
```

```python
movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
```

```python
movies_watched_df = user_movie_df[movies_watched]
```

```python
movies_watched_df.shape
```

```python
user_movie_count = movies_watched_df.T.notnull().sum()
```

```python
user_movie_count = user_movie_count.reset_index()
```

```python
user_movie_count.columns = ['userId', 'movie_count']
```

```python
perc = len(movies_watched) * 60 / 100
```

```python
users_same_movies = user_movie_count[user_movie_count['movie_count'] > perc]['userId']
```

```python
final_df = pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movies)],
                     random_user_df[movies_watched]])
```

```python
final_df.head()
```

```python
corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
```

```python
corr_df = pd.DataFrame(corr_df, columns=['corr'])
corr_df.index.names = ['user_id_1','user_id_2']
corr_df = corr_df.reset_index()
corr_df.head()
```

```python
top_users = corr_df[(corr_df['user_id_1'] == 108170) & 
                    (corr_df['corr'] > 0.65)][['user_id_2','corr']].reset_index(drop=True)
```

```python
top_users = top_users.sort_values(by='corr', ascending=False)
```

```python
top_users.rename(columns={'user_id_2':'userId'}, inplace=True)
```

```python
top_users_ratings = top_users.merge(rating[['userId','movieId','rating']], how='inner')
```

```python
top_users_ratings = top_users_ratings[top_users_ratings['userId'] != 108170]
```

```python
top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
```

```python
top_users_ratings.head()
```

```python
recommendation_df = top_users_ratings.groupby('movieId').agg({'weighted_rating':'mean'})
recommendation_df = recommendation_df.reset_index()
```

```python
recommendation_df.sort_values(by='weighted_rating', ascending=False)
```

```python
recommendation_df['movieId'].nunique()
```

### Item-Based Recommendation

```python
def create_user_movie_df():
    movie = pd.read_csv('/Users/yelizsari/Downloads/VBO/6. VBO - Recommendation/movie.csv')
    rating = pd.read_csv('/Users/yelizsari/Downloads/VBO/6. VBO - Recommendation/rating.csv')
    df = movie.merge(rating, how='left', on='movieId')
    rating_counts = pd.DataFrame(df.title.value_counts())
    rare_movies = rating_counts[rating_counts['title'] <= 1000].index
    common_movies = df[~df['title'].isin(rare_movies)]
    return common_movies
```

```python
user_movie_df = create_user_movie_df()
```

```python
user_movie_df[user_movie_df['userId']==108170].sort_values(by='timestamp', ascending=False)
```

```python
user_movie_df = user_movie_df.pivot_table(index=['userId'], columns=['title'], values='rating')
```

```python
def item_based_recommender(movie_name, user_movie_df):
    movie_name = user_movie_df[movie_name]
    return user_movie_df.corrwith(movie_name).sort_values(ascending = False)
```

```python
movies_from_item_based = item_based_recommender('Wild at Heart (1990)', user_movie_df)
```
